```python
import cv2
import sys
import numpy as np
import matplotlib.pyplot as plt

src = cv2.imread('lenna.bmp', cv2.IMREAD_GRAYSCALE)
```

cv2.calcHist(images, channels, mask, histSize, ranges, hist = False, accumulate = False)

images 는 입력 영상 리스트


channels 는 히스토그램 구할 색상 채널을 나타냄 

영상이 그레이 스케일이면 channel = [0] 그레이 스케일 1개

컬러 영상이면 [0,1] BGR 중 blue, green 2개 히스토 그램 ndarray로 반환 


mask는 영상중 특정부분에 대한 연산을 진행할 때 지정해 주는것


histsize는 히스토그램 차원의 크기를 나타낸다 256이 최대,


128이면 0또는1인것, 2또는3인것 이런식으로 구분되어 히스토그램 출력

ranges는 컬러 영상의 경우 각채널의 최소 최대 범위정해줌 [0, 256, 0, 256, 0, 256] 


```python
if src is None:
    print('Image load failed!')
    sys.exit()

hist = cv2.calcHist([src], [0], mask = None, histSize = [256], ranges = [0, 256])

cv2.imshow('src_gray', src)


plt.plot(hist)
plt.show()

src = cv2.imread('lenna.bmp', cv2.IMREAD_COLOR)

if src is None:
    print('Image Load failed!')
    sys.exit()

colors = ['b', 'g', 'r']
bgr_planes = cv2.split(src)


for (p, c) in zip(bgr_planes, colors):
    hist = cv2.calcHist([p], [0], None, [256], [0, 256])
    plt.plot(hist, color = c)

plt.show()


cv2.imshow('src_color', src)

if cv2.waitKey() == 27:
    cv2.destroyAllWindows()
 


```
