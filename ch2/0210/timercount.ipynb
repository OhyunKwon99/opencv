```python
import sys
import cv2
import numpy as np
import time
import math
```

tm: cv2.TickMeter 객체

tm.start(): 시간측정 시작

tm.stop(): 시간측정 끝

각각 특정부분의 시작지점, 끝지점에서 호출해주면 시간값 계산

tm.getTimeMilli(): m초 단위로 반환

```python
img = cv2.imread('hongkong.jpg')

fps = [4]

if img is None:
    print('Image load failed!')
    sys.exit()


def arr_count (arr):
    cnt = 0
    for x in arr:
        x != 0.0
        cnt = cnt + 1
    return float(cnt)

cap = cv2.VideoCapture(0,cv2.CAP_DSHOW)
##cap.set(cv2.CAP_PROP_FRAME_WIDTH , 1280)
##cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 960)

while True:
    tm = cv2.TickMeter()
    tm.start()
    status, frame = cap.read()

    if not status: 
        break

    edge = cv2.Canny(frame, 50, 150)
    
    cv2.imshow('Normal', frame)
    cv2.imshow('Canny edge', edge)

    tm.stop()
    sec = tm.getTimeSec()
    fps.append(sec)

    if len(fps) >= 5:
        fps.pop(0)

    mean_sec = np.sum(fps, dtype = float)/arr_count(fps)
    print('Elapsed Time: {}(s), Mean Elapsed Time: {}(s)'.format(sec, mean_sec) )
    print('cal_FPS : {}'.format(1.0/mean_sec))
    print('get_FPS : {}'.format(cap.get(cv2.CAP_PROP_FPS)))
    if cv2.waitKey(1) == 27:
        break

cap.release()
cv2.destroyAllWindows()



    







```
