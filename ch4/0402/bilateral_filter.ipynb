가우시안 필터는 모든 부분이 convolution되면서 전체적인 blur가 되버림

양방향 필터를 통해 edge가 아닌 부분에서만 modified된 가우시안 블러링 사용

sigma color 값을 통해 엣지부분의 픽셀변화량 기준치를 정해주고 그 값을 기준으로 edge를 판단하게 됨

단 필터를 상황마다 수정해서 작용하기 때문에 일반 가우시안 필터보다 시간이 많이 딜레이된다.

```python
import sys
import cv2
import numpy as np
from matplotlib import pyplot as plt

src = cv2.imread('lenna.bmp')

if src is None:
    print('Image load failed')
    sys.exit()

dst = cv2.bilateralFilter(src, (-1), 10, 5)

title = ['src', 'dst']

def BGR2RGB(source, destination):
    src = cv2.cvtColor(source, cv2.COLOR_BGR2RGB)
    dst = cv2.cvtColor(destination, cv2.COLOR_BGR2RGB)
    for i in range(2):
        plt.subplot(1, 2, i+1)
        plt.imshow(src)
        plt.imshow(dst)
        plt.title(title[i])
    plt.show()
   
BGR2RGB(src, dst)   

cv2.imshow('src', src)
cv2.imshow('dst', dst)

cv2.waitKey()
cv2.destroyAllWindows()
```
