```python
import matplotlib.pyplot as plt
import cv2


# 빈 컬러 영상 출력
imgBGR = cv2.imread('test.bmp')
imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)


##그레이 스케일 영상 출력
imgGray = cv2.imread('test.bmp', cv2.IMREAD_GRAYSCALE)


##두개의 영상이나 사진 한창에 출력하는법
plt.subplot(121), plt.axis('off'), plt.imshow(imgRGB)
plt.subplot(122), plt.axis('off'), plt.imshow(imgGray, cmap = 'gray')
#plt.axis('off') #눈금 없애는 용도
plt.show()
```

1. 그레이 스케일로 imGray에 2차원 행렬 형태로 저장 ndarray
2. plt.imshow(imgGray, cmap = 'gray') 은 cmap 언급 필수
